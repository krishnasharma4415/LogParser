```python
def read(folder_path):
  log_arrays = {}
  for filename in os.listdir(folder_path):
    if filename.endswith('.log') or filename.endswith('.csv'):
      with open(os.path.join(folder_path, filename), 'r') as f:
        log_array = f.readlines()
        log_name = filename[:-7]
        log_arrays[log_name] = log_array
        print(log_name)
  return log_arrays

raw_folder_path = "../raw_data"
raw_logs = read(raw_folder_path)
```

```python
def read(folder_path):
  log_arrays = {}
  for filename in os.listdir(folder_path):
    if filename.endswith('.log') or filename.endswith('.csv'):
      with open(os.path.join(folder_path, filename), 'r') as f:
        log_array = f.readlines()
        log_name = filename[:-22]
        log_arrays[log_name] = log_array
        print(log_name)
  return log_arrays

structured_folder_path = "../structured_data_labelled"
structured_logs = read(structured_folder_path)
```

```python
columns = ["LineId", "Date", "Time", "Level", "Component", "PID", "Content", "EventId", "EventTemplate", "Threat"]
```

```python
def unify_schema(df, columns):
    for col in columns:
        if col not in df.columns:
            df[col] = 'NA' if col not in ["LineId", "PID"] else 0
    return df[columns]

df = pd.read_csv('../structured_data_labelled/OpenSSH_2k.log_structured.csv')
unify_df = unify_schema(df, columns)

# structured_logs_dfs = {}
# for log_name, log_data in structured_logs.items():
#     from io import StringIO
#     log_data_str = ''.join(log_data)  
#     df = pd.read_csv(StringIO(log_data_str))  
    
#     df = unify_schema(df, columns)
#     structured_logs_dfs[log_name] = df
```

```python
from datetime import datetime

def standardize_datetime(df):
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%m-%d')
    df['Time'] = pd.to_datetime(df['Time'], errors='coerce').dt.strftime('%H:%M:%S')
    return df

android_logs = standardize_datetime(unify_df)
```
